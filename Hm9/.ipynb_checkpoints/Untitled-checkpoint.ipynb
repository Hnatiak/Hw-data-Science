```python
import tensorflow as tf
```

Завантажуємо набір даних Fashion MNIST

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

Імпортуємо бібліотеку для побудови графіків

```python
import matplotlib.pyplot as plt
```

Відображаємо перше зображення з тренувального набору

```python
plt.imshow(X_train[0])
```

Проведемо нормалізацію даних, розділяючи значення кожнрго пікселя на 255

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

Визначаємо функцію, яка прийматиме гіперпараметри (функцію активації, кількість нейронів у шарі та швидкість навчання) і повертатиме скомпільовану модель нейронної мережі.

```python
def build_model(hp):
  model = tf.keras.Sequential() # Створюємо порожню послідовну модель Keras
  model.add(tf.keras.layers.Input(shape=(28, 28))) # Додаємо перший шар Flatten для вирівнювання вхідних даних (28x28 пікселів)
  # model.add(tf.keras.layers.Flatten(input_shape=(28, 28))) # Додаємо перший шар Flatten для вирівнювання вхідних даних (28x28 пікселів)
  hp_activation = hp.Choice('activation', values=['relu', 'swish', 'gelu']) # Вибираємо випадковим чином тип активації зі списку ['relu', 'swish', 'gelu']
  
  # Вибираємо випадковим чином кількість нейронів у першому прихованому шарі
  hp_layer_1 = hp.Int('layer_1', min_value=1, max_value=1000, step=100) # Значення вибирається в межах від 1 до 1000 з кроком 100
  hp_layer_2 = hp.Int('layer_2', min_value=1, max_value=1000, step=100) # Аналогічно вибираємо кількість нейронів у другому прихованому шарі
  hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4]) # Вибираємо випадковим чином швидкість навчання зі списку [0.01, 0.001, 0.0001]

  model.add(tf.keras.layers.Flatten())  # Вирівнюємо вхідні дані перед Dense шарами
    
  model.add(tf.keras.layers.Dense(units=hp_layer_1, activation=hp_activation)) # Додаємо перший прихований шар з вибраною активацією та кількістю нейронів
  model.add(tf.keras.layers.Dense(units=hp_layer_2, activation=hp_activation)) # Додаємо другий прихований шар з вибраною активацією та кількістю нейронів
  model.add(tf.keras.layers.Dense(10, activation='softmax')) # Додаємо останній шар з функцією активації softmax для класифікації на 10 класів
    
  # Компілюємо модель з оптимізатором Adam, обраною швидкістю навчання, функцією втрати SparseCategoricalCrossentropy
  # для класифікації категорій, та метрикою точності (accuracy)
  model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate),
                loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                metrics=['accuracy'])

  # Повертаємо побудовану модель
  return model
```

Створимо об'єкт Keras Tuner, використовуючи алгоритм Hyperband для пошуку найкращих гіперпараметрів.

Тут ми вказуємо функцію побудови моделі, фактор зменшення, максимальну кількість епох

Тому для цього імпортнемо нашу бібліотеку яка допоможе нам у цьому: keras_tuner

```python
import keras_tuner as kt
```

Створюємо об'єкт tuner для гіперпараметричного пошуку моделі за допомогою Hyperband

```python
tuner = kt.Hyperband(build_model,
                     objective=kt.Objective("val_accuracy", direction="max"),  # Визначаємо метрику для максимізації (точність на валідаційній вибірці)
                     max_epochs=10,  # Максимальна кількість епох для пошуку
                     factor=3,  # Фактор зменшення кількості моделей у Hyperband
                     directory='model',  # Директорія для збереження результатів
                     project_name='fashion_mnist')  # Назва проекту (наприклад, набір даних fashion_mnist)
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=3)
```

Виконуємо пошук оптимальних гіперпараметрів

Тут ми вказуємо дані для еавчання, кількість епох, співвідношення валідаційного набору і також використовуємо методі ранньої ззупинки

```python
tuner.search(X_train, y_train, epochs=10, validation_split=0.2, callbacks=[stop_early])
```

Отримуємо найкращі гіперпараметри, що були знайдені під час пошуку

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

Створюємо модкль із найліпшими гіперпараметрами і приступаємо до її навчання на тренувальних даних

```python
model = tuner.hypermodel.build(best_hps)
history = model.fit(X_train, y_train, epochs=30, validation_split=0.2, callbacks=[stop_early])
```

Перегляньмо метрики для показання ефективності нейронної мережі:

<ul>
    <li>Графік точності на тренувальному та валідаційному наборах</li>
</ul>

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()
```

<ul>
    <li>Графік втрат на тренувальному та валідаційному наборах</li>
</ul>

```python
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label = 'Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

<ul>
    <li>Графік точності та втрат на тестовому наборі</li>
</ul>

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['loss'], label='Train Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy / Loss')
plt.title('Training Accuracy and Loss')
plt.legend()
plt.show()
```

<h3>Підведемо підсумок оцінки нашої натренованої моделі і виведемо значення втрат і точності моделі</h3>

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
```
