Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouping_columns=['Посада']
columns_to_show=["Зарплата на місяць"]
grouped_df=python_data.groupby(grouping_columns)[columns_to_show].agg([np.min, np.max])
grouped_df
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
grouping_columns=['Посада']
columns_to_show=["Зарплата на місяць"]
grouped_df=python_data.groupby(grouping_columns)[columns_to_show].agg([np.mean, np.min, np.max])
grouped_df=grouped_df.sort_values(by=('Зарплата на місяць', 'mean'))
grouped_df.plot(kind='bar', label=['average', 'min', 'max'])
plt.legend()
plt.show()
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return python_data[python_data['Посада']==row['Посада']]["Зарплата на місяць"].mean()

fill_avg_series=python_data.apply(fill_avg_salary, axis=1)
python_data.insert(loc=len(python_data.columns), column='Середня зарплата', value=fill_avg_series)
python_data
```

```python
pd.Categorical(python_data['Рівень англійської'], categories=['елементарний', 'нище середнього', 'середній', 'вище середнього', 'просунутий'], ordered=True)
plt.figure(figsize=(8, 4))
sns.swarmplot(x='Рівень англійської', y='salary', hue='Посада', data=python_data)
plt.title('Зарплата в залежності від знань англійської мови')
plt.xlabel('Рівень англійської мови')
plt.ylabel('Зарплата')
plt.show()
```

```python
python_data['Загальний досвід роботи'] = python_data['Загальний досвід роботи'].astype(str)
python_data['Загальний досвід роботи'].replace(to_replace="10 і більше років", value="10", inplace=True)
python_data['Загальний досвід роботи'].replace(to_replace="менше 3 місяців", value="0", inplace=True)
python_data['Загальний досвід роботи'] = python_data['Загальний досвід роботи'].astype(float)
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Загальний досвід роботи', y='salary', data=python_data)
plt.title('Зарплата в залежності від досвіду роботи')
plt.xlabel('Досвід роботи і роки')
plt.ylabel('Зарплата')
plt.show()
```
